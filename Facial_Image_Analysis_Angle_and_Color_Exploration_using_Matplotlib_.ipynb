##matplotlib.pyplot

is a collection of command style functions that make Matplotlib work like MATLAB. Each Pyplot function makes some change to a figure. For example, a function creates a figure, a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

(See [Reference to different types of Plots](https://www.tutorialspoint.com/matplotlib/matplotlib_pyplot_api.htm))

(Ref - TutorialPoint)

```python
import matplotlib.pyplot as plt # this is the default declaration
# we are importing from the matplotlib package, pyplot is a class and
# we are aliasing it as plt

import numpy as np # We need numpy as well
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05) # generate an array from 0 to math.pi*2 with
                                  # an interval of 0.05
y = np.sin(x)  # built in trignometry function in numpy
plt.plot(x,y)  # an x, y plot
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show() # the plot is shown only when you give the show command
```

Try to plot y=3x2+5x+10

You can get different types of colours and line styles for your plot see Line Styles and Colours

```python
#simple plot graph
import matplotlib.pyplot as plt
import numpy as np

# Define the function y = 3x^2 + 5x + 10
def f(x):
    return 3*x**2 + 5*x + 10

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)
# Calculate y values
y = f(x)

# Plot the function with different line styles and colours
plt.figure(figsize=(8,6))
plt.plot(x, y, 'r--', label='Red dashed')   # Red dashed line
plt.plot(x, y+50, 'g-.', label='Green dash-dot')  # Green dash-dot line, shifted up for visibility
plt.plot(x, y-50, 'b:', label='Blue dotted')       # Blue dotted line, shifted down for visibility

plt.title('Plot of y=3x^2 + 5x + 10 with Different Line Styles and Colours')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

SubPlots
Matplotlibs pyplot API has a convenience function called subplots() which acts as a utility wrapper and helps in creating common layouts of subplots, including the enclosing figure object, in a single call.

Plt.subplots(nrows, ncols) The two integer arguments to this function specify the number of rows and columns of the subplot grid. The function returns a figure object and a tuple containing axes objects equal to nrows*ncols. Each axes object is accessible by its index. Here we create a subplot of 2 rows by 2 columns and display 4 different plots in each subplot.

(Ref - TutorialPoint)

```python
import matplotlib.pyplot as plt
fig,axes =  plt.subplots(2,3) # subplots(rows, columns), this function returns
                              # two values, fig and axes, we are interested
                              # in axes

import numpy as np
x = np.arange(1,5,0.2)        # an array with values 1, to 5 with 0.2 steps
axes[0][0].plot(x,x*x)        # first plot function at (0,0)
axes[0][0].set_title('square')
axes[0][1].plot(x,np.sqrt(x))  # plot at (0,1)
axes[0][1].set_title('square root')
axes[0][2].plot(x,np.exp(x))  # plot at (0,2)
axes[0][2].set_title('exp')
axes[1][0].plot(x,np.log10(x)) # plot at (1,0)
axes[1][0].set_title('log')
axes[1][1].plot(x, 2*x + 5)    # plot at (1,1)
axes[1][1].set_title("Linear")
axes[1][2].plot(x, 2*x**2 - 11*x + 10) # plot at (1,1)
axes[1][2].set_title("Polynomial")
plt.tight_layout()   # spaces out the plots

plt.show() # display all the plots
```

Seperating an image to 3 colour channels

```python
# Let's try to print the Obama image using the three colour channels seperately
# Lets try another image
import matplotlib.pyplot as plt
import matplotlib.image as img # take image class of matplotlib as img
import numpy as np
image = img.imread('obama.jpg') # please load this image to your folder
fig,axes =  plt.subplots(2,3,figsize = (7,7))

# We can take the different colour channels

# cmap are colour maps see https://matplotlib.org/tutorials/colors/colormaps.html

axes[0][0].imshow(image[:,:,0],cmap='Reds') # Red Channel
axes[0][1].imshow(image[:,:,1],cmap='Greens') # Green Channel
axes[0][2].imshow(image[:,:,2],cmap='Blues') # Blue Channel
axes[1][0].imshow(image[:,:,0],cmap='gray') # Red Channel in gray colour
axes[1][1].imshow(image[:,:,1],cmap='gray') # Green Channel in gray colour
axes[1][2].imshow(image[:,:,2],cmap='gray') # Blue Channel in gray colour

plt.tight_layout()   # spaces out the plots

plt.show() # display all the plots
```

```python
# Let's try to see images in one of the folders from the Georgia dataset
import matplotlib.pyplot as plt
import matplotlib.image as img # take image class of matplotlib as img
import numpy as np
fig,axes =  plt.subplots(3,5, figsize = (12,5))
sample = "s10" # change this to get a different folder
images = np.empty([15, 480, 640,3], dtype=float)
for r in range(15):
  if (r+1) < 10:
     zero = "0" # to check if we need to have a trailing zero
  else:
     zero = ""
  st = 'gt_db/'+sample+'/'+zero+str(r+1)+'.jpg' # generate file name
  print(st) # display loaded file as debug
  image = img.imread(st) # load image file
  images[r] = image

  y = int(r / 5) # calculate position in plot
  x = r % 5 # calculate position in plot
  axes[y][x].imshow(image) # set the image

plt.tight_layout()   # spaces out the plots
plt.show() # display all the plots

```

y=x*x Graph


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function y = x^2
def my_function(x):
    return x**2

# Generate x values and calculate y values
x = np.linspace(-5, 5, 11)  # 11 points from -5 to 5
y = my_function(x)

# Plot the function curve
plt.plot(x, y, label='y = x^2', color='blue')

# Plot the data points as red dots
plt.scatter(x, y, color='red')

# Annotate numeric values on the plot
for (i, j) in zip(x, y):
    plt.text(i, j, f'({i:.1f}, {j:.1f})', fontsize=8, ha='right', va='bottom')

plt.title('Plot of y = x^2 with Data Points and Numeric Annotations')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
